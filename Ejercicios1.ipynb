Ejercicio 1: Cree	1000	observaciones	de	la	serie	temporal	x(t)=4*x(t-1)*(1-x(t-1)) comenzando	en	
x(0)=0.1

```python
import numpy as np
serie = []
serie.append(.1)
import matplotlib.pyplot as plt
for i in range(1_000):
  anterior= serie[-1]
  serie.append(4 * anterior * (1 - anterior))
print(serie)
```

Ejercicio 2: Dibuje	su	evolución	partiendo	de	x(0)=0.1.	Tiene	algún	patrón	dinámico	apreciable?	
(por	ejemplo,	la	función	seno	muestra	regularidad	con	frecuencia	2*pi)

```python
plt.figure(figsize= (15, 10))
plt.plot(serie);
```

Ejercicio 3: Construya	un	modelo	x(t)=a	+	bx(t-1)	utilizando	la	ecuación	normal	.	El	valor	del	
parámetro	b	es	significativo?	(intuitivamente,	sin	emplear	conceptos	estadísticos)

```python
from scipy.ndimage.interpolation import shift 
y=serie
Xant=shift(serie,1)
Xone=np.ones(len(Xant))
X=np.vstack((Xant,Xone)).T

param= np.linalg.inv(X.T @ X) @ (X.T @ y)
print(param)
```

Ejercicio 4:	Evalúe	la	baja	capacidad	explicativa	del	modelo,	por	ejemplo	tratando	de	predecir	
x(t)	usando	el	modelo	lineal.	Considerando	esto,	a	su	juicio	los	resultados	obtenidos		
son	debidos:

```python
ypred=X@param
ynaive=0.5*np.ones(len(y))
naepred=np.abs(y-ypred)
naenaive=np.abs(y-ynaive)
print(np.sum(naepred))
print(np.sum(naenaive))
```
